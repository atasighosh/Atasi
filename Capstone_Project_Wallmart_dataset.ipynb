**Predictive Modelling to forcast the weekly_sales of each store for next 12 months**

```python
# For store 12 prediction and forcast has been done
# same procedure can be followed for any store
```

```python
store12_data = df[df['Store'] == 12]

store12_data.head()
```

**Modelling**

```python
df1=pd.DataFrame(store12_data[["Date","Weekly_Sales"]])
df1.head()
```

```python
from datetime import datetime
#%matplotlib inline
df2=df1.copy()
df2.set_index(["Date"],inplace=True)
df2.plot(figsize=(10,5))
```

```python
mean_log = df2.rolling(window=52).mean()  #12 month averge(mean)
std_log = df2.rolling(window=52).std()    #12 month devaiation from mean or (std)
df2.plot( color='blue', label='Original',figsize=(10,5))
plt.plot(mean_log, color='red', label='Rolling Mean')
plt.plot(std_log, color='black', label='Rolling Std')
plt.legend(loc='best')
plt.title('Rolling Mean & Standard Deviation (log-log Scale)')
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition=seasonal_decompose(df2[["Weekly_Sales"]], period=12)
print(decomposition.plot())
```

**ADF test**

```python
from statsmodels.tsa.stattools import adfuller
results=adfuller(df2[["Weekly_Sales"]])
p_value=results[1]
if p_value <0.05:
  print("p_value:",p_value,"\ndata is stationary")
```

**ACF Plot**

```python
from statsmodels.tsa.stattools import acf
from statsmodels.graphics.tsaplots import plot_acf

acf_plot=acf(df2[["Weekly_Sales"]].dropna())
print(plot_acf(acf_plot))
```

**PACF test**

```python
from statsmodels.tsa.stattools import pacf
from statsmodels.graphics.tsaplots import plot_pacf

pacf_plot=pacf(df2[["Weekly_Sales"]].dropna())
print(plot_pacf(pacf_plot,lags=10))
```

```python
df3=df2.reset_index(drop=True)
train=df3.iloc[:105][["Weekly_Sales"]]
test=df3.iloc[105:][["Weekly_Sales"]]
```

```python
 print(len(train))
 print(len(test))
```

```python
# expt with random p,d,q values
from statsmodels.tsa.arima.model import ARIMA  #autoregressive intergrated moving averge
model=ARIMA(train,order=(2,1,3)) ;  #(p,d,q),p=autoregressive,d=differencing,q=moving average
import warnings
warnings.filterwarnings("ignore")
model=model.fit()
df3["predict"]=model.predict(start=len(train),end=len(train)+len(test)-1)
df3.tail()
```

**Itertools for p,d,q values**

```python
import itertools
p=range(4)
d=range(0,2)
q=range(4)

pdq_combination=list(itertools.product(p,d,q))
pdq_combination
```

```python
from sklearn.metrics import mean_squared_error
rmse=[]
order1=[]
for pdq in pdq_combination:
  model=ARIMA(train,order=pdq)
  model_fit=model.fit()
  pred=model_fit.predict(start=len(train),end=len(train)+len(test)-1)
  error=np.sqrt(mean_squared_error(test,pred))
  order1.append(pdq)
  rmse.append(error)
  results=pd.DataFrame(index=order1,data=rmse,columns=["RMSE"])
  results.sort_values("RMSE")
```

**Forecasting**

**SARIMAX MODEL**

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
model=SARIMAX(train,order=(2,1,0),seasonal_order=(2,1,0,4))
model=model.fit()
df5["predict"]=model.predict(start=len(train),end=len(train)+len(test)-1)
ax=df5[["Weekly_Sales","predict"]].plot()
ax.set_xlabel("Index")
ax.set_ylabel("Sales")
```

```python
forecast = model.forecast(steps=120)
df5[["Weekly_Sales","predict"]].plot()
forecast.plot()
```
